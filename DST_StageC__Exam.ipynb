```
consumpDataset = pd.read_csv('Africa Food Supply (2004 - 2013) (1).csv')
consumpDataset
```

```
p = prodDataset.drop('Item', axis = 1)
totprod = p.groupby(['Country','Year'])['Value'].sum()
totprod = pd.DataFrame(totprod).reset_index()
totprod
```

```
import plotly.express as px
fig = px.line(totprod, x=totprod['Country'], y=totprod['Value'], color=totprod['Year'])
fig.show()
```

```
code = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/2014_world_gdp_with_codes.csv')
assign_code = code.drop('GDP (BILLIONS)', axis=1)
new_name = {'COUNTRY':'Country'}
assign_code = assign_code.rename(columns = new_name)
production = pd.merge(totprod, assign_code, on = 'Country')
production
```

```
fig = px.choropleth(production, locations='CODE', hover_name='Country', title='Food Production Trend(2004 -2013)', color='Value', scope="africa", labels={'Value':'Food Produced'}, color_continuous_scale="Plasma", animation_frame="Year")
fig.show()
```

```
print('Nigeria, Egypt and South Africa, produces the most food across Africa')
```

```
totsupply = consumpDataset.groupby(['Country','Year'])['Value'].sum()
totsupply = pd.DataFrame(totsupply).reset_index()
totsupply
```

```
fig = px.line(totsupply, x=totsupply['Year'], y=totsupply['Value'], color=totsupply['Country'])
fig.show()
```

```
supply = pd.merge(consumpDataset, assign_code, on = 'Country')
supply
```

```
fig = px.choropleth(supply, locations='CODE', hover_name='Country', title='Food Supply/Consumption Trend(2004 -2013)', color='Value', scope="africa", labels={'Value':'Food Supplied'}, color_continuous_scale="Plasma", animation_frame="Year")
fig.show()
```

```
print('The trend shows that Egypt, Tunisia and Morocco foods production is adequate for the their supply/consumption demands. While countries like Nigeria and South Africa that has the highest trends(along with Egypt) in food production but not adequate enough for their countries supply/consumption demands.')
```

```
import numpy as np
sort_prod = np.sort(prodDataset['Value'])
Q1 = np.percentile(prodDataset['Value'], 25, interpolation = 'midpoint')
Q2 = np.percentile(prodDataset['Value'], 50, interpolation = 'midpoint')
Q3 = np.percentile(prodDataset['Value'], 75, interpolation = 'midpoint')
IQR = Q3 - Q1
print('Q1:', Q1)
print('Q2:', Q2)
print('Q3:', Q3)
print('Interquartile Range:', IQR)
```

```
low_lim = Q1 - 1.5 * IQR
up_lim = Q3 + 1.5 * IQR
print('Lower Limit:', low_lim)
print('Upper Limit:', up_lim)
```

```
production_average = prodDataset.groupby('Year')['Value'].mean()
production_average
```

```
consumption_average = consumpDataset.groupby('Year')['Value'].mean()
consumption_average
```

```
from scipy.stats import pearsonr
stat, p = pearsonr(consumption_average, production_average)
print('Pearsons correlation: Statistics: %.3f' % stat, 'P-Value:' % stat, p)
if p > 0.05: print('H0, accept null hypothesis; Not Significant')
else: print('H1, reject null hypothesis; Significant')
```

```
import plotly.graph_objects as go
fig = go.Figure()
fig.add_trace(go.Scatter(x=consumption_average, y=production_average, name='Average Food Consumption vs Average Food Production(2004 - 2013)'))
fig.show()
```
